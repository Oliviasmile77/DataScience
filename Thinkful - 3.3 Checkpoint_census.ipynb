```python
import pandas as pd

state_populations = pd.read_csv(r'C:\Users\ojiang1\Desktop\Data Science\state_populations.csv')
census_divisions = pd.read_csv(r'C:\Users\ojiang1\Desktop\Data Science\census_divisions.csv')

census = pd.merge(state_populations, census_divisions)
census.head(5)
```
